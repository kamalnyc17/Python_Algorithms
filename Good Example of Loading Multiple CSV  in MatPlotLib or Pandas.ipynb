```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from functools import reduce
import os
```

```python
csv_files = []
for filename in sorted(os.listdir("../Resources")):
    if filename.endswith(".csv"):
        csv_files.append(filename)
```

```python
years = []
dfs = []
for file in csv_files:
    df = pd.read_csv(f'../Resources/{file}')
    dfs.append(df)
    year = file.split("-")
    years.append(year[2].split(".")[0])
```

```python
var=iter(["2013", "2014", "2015", "2016"])
var1=iter(["2014", "2015", "2016"])

var=iter(years)
years2 = years.copy()
del years2[0]
var1=iter(years2)
df_final_combined = reduce(lambda left,right: pd.merge(left, right, how='outer', on='Wrestler',suffixes=[next(var),next(var1)]),dfs)
df_final_combined.head(10)
```

```python
# Replace all NaN values with 0 
combined_wrestlers_df = df_final_combined.fillna(0)

def get_col_name(x):
    return [i for i in list(combined_wrestlers_df.columns) if x in i]


wins_col = get_col_name('Wins')
loss_col = get_col_name('Losses')
draw_col = get_col_name('Draws')

# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values
combined_wrestlers_df["Total Wins"] = combined_wrestlers_df.loc[:, wins_col].sum(axis=1)
# # Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values
combined_wrestlers_df["Total Losses"] = combined_wrestlers_df.loc[:, loss_col].sum(axis=1)

# # Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values
combined_wrestlers_df["Total Draws"] = combined_wrestlers_df.loc[:, draw_col].sum(axis=1)

total_col = get_col_name('Total')

combined_wrestlers_df["Total Matches1"] = combined_wrestlers_df.loc[:, total_col].sum(axis=1)

combined_wrestlers_df
```
